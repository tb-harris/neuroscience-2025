## Process of creating a visualization

When creating a visualization, **start with the most basic code possible**- for example, sns.histplot() with minimal parameters

For example, start with something like this (**no** axis labels, scaling, bin selection, hue, etc).

Start with the basics, and then improve from there.

```python
sns.histplot(x=rna_data['Creb1'])
```

Once you have the most basic figure, then think about how you can improve it:

```python
sns.histplot(x=rna_data['Creb1'])
plt.xlabel("Creb1 Expression (cpm)")
plt.ylabel("Cell Count")
```

With histograms, we can then examine different configurations of bin, binwidth, binrange parameters: see our visualization notes

```python
sns.histplot(x=rna_data['Creb1'], binwidth=20)
plt.xlabel("Creb1 Expression (cpm)")
plt.ylabel("Cell Count")
```

For histograms, it can also be helpful to show things in relative rather than absolute terms:

```python
sns.histplot(x=rna_data['Creb1'], stat="percent", binwidth=20)
```

Now, we can add our labels back in- remember to look at the default / most basic figure first!

```python
sns.histplot(x=rna_data['Creb1'], stat="percent", binwidth=20)
plt.ylabel("% of cells")
plt.xlabel("Creb1 expression (cpm)")
```

When adding our graph to a slide, we should also add an appropriate **caption** or title (if you don't add one in python) describing the graph

## Axis scale/range

We can put histograms on the same scale by using `sns.displot()` (see [previous notes](https://github.com/tb-harris/neuroscience-2025/blob/main/1_Visualizing_and_Describing_Distributions.ipynb)).

We can also set the scale of plots directly by using `plt.xlim()` and `plt.ylim()`. **Make sure to start without limiting your axes, and ensure that you aren't excluding data points unless you intend to and clearly indicate what is truncated in your figure title/caption.**

```python
sns.histplot(x=rna_data['Creb1'], binwidth=20)
plt.ylim(top=2000)
```
